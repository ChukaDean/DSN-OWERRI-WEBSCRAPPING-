```python
import requests
from bs4 import BeautifulSoup
page = requests.get ("https://www.soccerbase.com/teams/head_to_head.sd?team_id=1718&team2_id=142")
#page = requests.get ("https://www.uusuuiiwii87890.com")
if page.status_code == 200:
    print ("Successful")
elif page.status_code == 404:
    print ("Not Found")

soup = BeautifulSoup(page.content, "html.parser")
print (soup)
```

```python
Date_Time = soup.find_all(class_ = "dateTime")
#print (Date_Time [2])
#print (Date_Time [2].text)
Date_Time_Soccer= [yt.get_text(strip = True) for yt in Date_Time]
print(Date_Time_Soccer)
```

```python
League = soup.find_all(class_ = "first tournament")
#print (League)
Soccer_League = [sl.get_text(strip = True) for sl in League ]
print(Soccer_League)
```

```python
home_team = soup.find_all(class_ = "team homeTeam")
Home_Team = [pt.get_text() for pt in home_team]
#print (Home_Team[2])
print (Home_Team)
```

```python
import string
Scores = soup.find_all (class_ ="score")
#print (Scores)
Football_Score = [st.get_text(strip = True) for st in Scores]
print (Football_Score) 
```

```python
Away = soup.find_all(class_ = "team awayTeam")
Away_Team = [at.get_text() for at in Away]
print (Away_Team)

```

```python
import pandas as pd
import numpy as np
df = pd.DataFrame()
Football_Predict= pd.DataFrame(
{
    "Date_Time": Date_Time_Soccer,
    "League": Soccer_League ,
    "Home_Team" : Home_Team,
    "Score": Football_Score,
    "Away_Team": Away_Team,
    
}
)

#print (Football_Predict)
Football_Predict["Score"]
Football_Predict[['Home_Score','Away_Score']] = Football_Predict.Score.str.split("-",expand=True,)
Football_Predict['Home_Dummy_Score'] = np.where(Football_Predict['Home_Score'] >= Football_Predict['Away_Score'],1,0)
Football_Predict['Away_Dummy_Score'] = np.where(Football_Predict['Away_Score'] >= Football_Predict['Home_Score'],1,0)

print(Football_Predict)
```

```python
Football_Predict.to_csv("Soccer_Data", encoding='utf-8', index=False)
Football_Predict.to_excel("Soccer_Data_Excel", engine = 'xlsxwriter', encoding='utf-8', index=False)



```
